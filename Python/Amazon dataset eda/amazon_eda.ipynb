```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(9,9))
plt.scatter(ds['product_name'],ds['actual_price'])
plt.xlabel('product_name')
plt.ylabel('actual_price')
plt.show()
```

```python
def bar_plot(variable):
    var=ds[variable]
    varValue=var.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index,varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print(f"{variable}:\n {varValue}")
```

```python
ds.columns
```

```python
column_plot=['product_name','category','discounted_price','actual_price','discount_percentage','rating']
for i in column_plot:
    bar_plot(i)
```

```python
plt.hist(ds['discount_percentage'])
plt.show()
```

```python
ds['discounted_price'] = ds['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
ds['actual_price'] = ds['actual_price'].replace('[₹,]', '', regex=True).astype(float)
ds['discount_percentage'] = ds['discount_percentage'].replace('%', '', regex=True).astype(float)
ds['rating'] = pd.to_numeric(ds['rating'], errors='coerce')
ds['rating_count'] = ds['rating_count'].replace(',', '', regex=True).astype(float)
ds['rating'].fillna(ds['rating'].median(), inplace=True)
ds['rating_count'].fillna(ds['rating_count'].median(), inplace=True)
print(ds.isnull().sum())
```

```python
ds[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].describe()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(ds['actual_price'], bins=30, color='green')
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(ds['rating'],bins=20,color='blue')
plt.title('Distribution of Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
top_categories=ds['category'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.boxplot(data=ds[ds['category'].isin(top_categories)], x='category', y='actual_price')
plt.title('Price Distribution')
plt.xlabel('Category')
plt.ylabel('Actual Price')
plt.xticks(rotation=45)
plt.show()
```

```python
top_categories = ds['category'].value_counts().head(10).index
plt.figure(figsize=(10, 5))
sns.boxplot(data=ds[ds['category'].isin(top_categories)], x='category', y='actual_price')
plt.xticks(rotation=45)
plt.title('Price Distribution Across Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Actual Price')
plt.show()
```
