```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
yp=np.array([2,5,4,23,23,2])
plt.plot(yp,'o:r')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
a=np.array([3,4])
b=np.array([4,9])
plt.subplot(1,2,1)
plt.plot(a,b)
plt.show()
```

```python
x=np.array([4,5,6,78,9,10])
y=np.array([33,4,44,33,2,10])
plt.scatter(x,y)
x=np.array([32,32,44,4,34,34,231,4])
y=np.array([43,5,3,4,34,32,213,3])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array(['a','b','c','d'])
y=np.array([3,5,4,2])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(['a','b','c','d'])
y=np.array([3,5,4,2])
plt.barh(x,y)
plt.show()
```
