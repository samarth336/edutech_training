```python
plt.figure(figsize=(9,3))
plt.scatter(df['Age'],df['Survived'])
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()
```

```python
def bar_plot(variable):
    var=df[variable]
    varValue=var.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index,varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print(f"{variable}:\n {varValue}")
```

```python
column_plot=['Survived','Pclass','Name','Sex','SibSp','Parch']
for i in column_plot:
    bar_plot(i)
```
