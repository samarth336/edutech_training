```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tips.columns
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker')
```

```python
import numpy as np
rs=np.random.RandomState(10)
d=rs.normal(size=100)
sns.histplot(d,kde=True,color='m')
```

```python
sns.lineplot(
    x='timepoint',
    y='signal',
    hue='region',
    style='event',
    data=fmri
)
```
